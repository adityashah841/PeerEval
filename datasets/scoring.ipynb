```python
import os
import re
```

```python
folders = os.listdir('Dataset for informativeness')
```

```python
folders
```

```python
labels = ['ABS', 'INT', 'RWK', 'PDI', 'DAT', 'MET', 'EXP', 'RES', 'TNF', 'ANA', 'FWK', 'OAL', 'BIB', 'EXT', 'APR', 'NOV', 'IMP', 'CMP', 'PNF', 'REC', 'EMP', 'SUB', 'CLA']
```

```python
ans=[]
```

```python
path = 'Dataset for informativeness/'
for folder in folders:
    files = os.listdir(path+folder)
    for file in files:
        f = open(path+folder+'/'+file, 'r')
        content = f.read()
        pattern = r"\[(.*?)\]"
        result = re.findall(pattern, content)
        for i in range(len(result)):
            if i%2==0:
                result[i]=result[i][1:]
        for i in range(len(result)):
            if ',' in result[i]:
                a=list(result[i].split(','))
                result.pop(i)
                result.insert(i, a[0])
                for j in range(1, len(a)):
                    result.append(a[j])
        for el in result:
            if el=='null' or el=='':
                result.remove(el)
        for i in range(len(result)):
            while result[i][0]=='[':
                result[i]=result[i][1:]
        for el in result:
            if '-' not in el:
                result.remove(el)
        d={}
        for label in labels:
            d[label]=0
        for res in result:
            if '-' in res:
                if res.split('-')[1]=='POS':
                    try:
                        d[res.split('-')[0]]=d[res.split('-')[0]]+1
                    except:
                        d[res.split('-')[0]]=1
                elif res.split('-')[1]=='NEG':
                    try:
                        d[res.split('-')[0]]=d[res.split('-')[0]]-1
                    except:
                        d[res.split('-')[0]]=-1
        ans.append(d)
```

```python
l=[len(a) for a in ans]
```

```python
for a in ans:
    for el in a.keys():
        if el not in labels:
            for d in ans:
                print(list(d.keys()))
                if a not in list(d.keys()):
                    a[el]=0
```
